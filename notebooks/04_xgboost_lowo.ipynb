# LOWO with XGBoost + RandomizedSearchCV (GroupKFold)

**Objective:** Optimize hyperparameters using **GroupKFold** (cv=10) and validate with LOWO.

**Algorithm:** XGBoostRegressor

**Structure:**
1. Load data
2. Optimization with GroupKFold (cv=10)
3. LOWO validation
4. Statistical analyses
5. Visualizations
6. Save results to `results/xgboost/`

**Note:** The hyperparameter sensitivity analysis is in notebook `11_xgboost_sensitivity_analysis.ipynb`.

---

## 1. Imports and Settings

```python
import json
import time
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path
from xgboost import XGBRegressor
from sklearn.model_selection import RandomizedSearchCV, GroupKFold
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
from scipy.stats import uniform, randint, wilcoxon

from sonic_ml_utils import plot_well_profile_and_hexabin, statistics

import warnings
warnings.filterwarnings('ignore')
```

```python
# ── Directories ─────────────────────────────────────────────────────────────
RESULTS_DIR     = Path('../results/xgboost')
METRICS_DIR     = RESULTS_DIR / 'metrics'
PREDICTIONS_DIR = RESULTS_DIR / 'predictions'
MODELS_DIR      = RESULTS_DIR / 'models'
FIGURES_DIR     = RESULTS_DIR / 'figures'

for d in [METRICS_DIR, PREDICTIONS_DIR, MODELS_DIR, FIGURES_DIR]:
    d.mkdir(parents=True, exist_ok=True)

print('Directories configured.')
for d in [METRICS_DIR, PREDICTIONS_DIR, MODELS_DIR, FIGURES_DIR]:
    print(f'  {d}')
```

## 2. Load Data

```python
df = pd.read_csv('../data/processed/wells_iqr_with_lithology.csv')

print(f'Data loaded')
print(f'Total samples : {len(df):,}')
print(f'Total wells   : {df["Well_Name"].nunique()}')
print(f'Columns       : {df.columns.tolist()}')
```

```python
# ── Global settings ─────────────────────────────────────────────────────────
FEATURES     = ['GR', 'RT90', 'RHOB', 'NPHI']
TARGET       = 'DT'
N_ITER       = 50
N_SPLITS     = 10
RANDOM_STATE = 42

wells = df['Well_Name'].unique()

print(f'Features      : {FEATURES}')
print(f'Target        : {TARGET}')
print(f'Wells         : {len(wells)}')
print(f'RandomSearch  : {N_ITER} iterations, GroupKFold cv={N_SPLITS}')
```

## 3. Hyperparameter Optimization (GroupKFold)

GroupKFold ensures that samples from the same well do not appear simultaneously in training and test sets during tuning, preventing data leakage between wells.

```python
xgb_param_dist = {
    'n_estimators'     : randint(100, 500),
    'max_depth'        : randint(1, 10),
    'learning_rate'    : uniform(0.01, 0.49),
    'subsample'        : uniform(0.1, 0.9),
    'colsample_bytree' : uniform(0.5, 0.5),
    'gamma'            : uniform(1, 10),
    'reg_alpha'        : uniform(1, 10),
    'reg_lambda'       : uniform(100, 800),
    'min_child_weight' : randint(1, 5)
}

xgb_model = XGBRegressor(
    objective='reg:squarederror',
    random_state=RANDOM_STATE,
    tree_method='hist',
    device='cuda'  # Change to 'cpu' if no GPU is available
)

print(f'Search space: {len(xgb_param_dist)} hyperparameters, {N_ITER} iterations')
```

```python
# ── GroupKFold configuration validation ───────────────────────────────────────
print('GroupKFold CONFIGURATION VALIDATION')
print('=' * 60)

n_wells   = df['Well_Name'].nunique()
n_samples = len(df)

print(f'Dataset     : {n_samples:,} samples | {n_wells} wells')
print(f'N_SPLITS    : {N_SPLITS}')
print(f'Min required: N_SPLITS <= n_wells ({N_SPLITS} <= {n_wells})')

# ── Check 1: n_splits <= n_groups ────────────────────────────────────────────
if N_SPLITS > n_wells:
    print(f'\n[FAIL] N_SPLITS={N_SPLITS} > n_wells={n_wells}. '
          f'Reduce N_SPLITS to at most {n_wells}.')
    raise ValueError('N_SPLITS exceeds number of wells. Fix before proceeding.')
else:
    print(f'[OK]  N_SPLITS={N_SPLITS} <= n_wells={n_wells}')

# ── Check 2: inspect fold distribution ───────────────────────────────────────
X_val  = df[FEATURES].values
y_val  = df[TARGET].values
grps   = df['Well_Name'].values
gkf    = GroupKFold(n_splits=N_SPLITS)

print(f'\n{"Fold":>5s}  {"Train wells":>12s}  {"Val wells":>10s}  '
      f'{"Train samples":>14s}  {"Val samples":>12s}  Val well names')
print('-' * 110)

fold_stats = []
for fold, (train_idx, val_idx) in enumerate(gkf.split(X_val, y_val, groups=grps), 1):
    train_wells = set(grps[train_idx])
    val_wells   = set(grps[val_idx])
    overlap     = train_wells & val_wells
    fold_stats.append({
        'fold'          : fold,
        'n_train_wells' : len(train_wells),
        'n_val_wells'   : len(val_wells),
        'n_train_samples': len(train_idx),
        'n_val_samples'  : len(val_idx),
        'overlap'        : len(overlap),
    })
    val_names = ', '.join(sorted(val_wells))
    print(f'{fold:>5d}  {len(train_wells):>12d}  {len(val_wells):>10d}  '
          f'{len(train_idx):>14,}  {len(val_idx):>12,}  {val_names}')

# ── Check 3: no overlap between train and validation wells ───────────────────
total_overlaps = sum(s['overlap'] for s in fold_stats)
print()
if total_overlaps > 0:
    print(f'[FAIL] {total_overlaps} well(s) appear in both train and validation '
          f'in at least one fold. GroupKFold is not working as expected.')
else:
    print(f'[OK]  No well appears in both train and validation in any fold.')

# ── Check 4: minimum wells per fold ──────────────────────────────────────────
min_val_wells = min(s['n_val_wells'] for s in fold_stats)
if min_val_wells < 1:
    print(f'[FAIL] At least one fold has 0 validation wells.')
else:
    print(f'[OK]  All folds have at least {min_val_wells} validation well(s).')

print()
print(f'Summary: {N_SPLITS} folds | '
      f'avg train wells={sum(s["n_train_wells"] for s in fold_stats)/N_SPLITS:.1f} | '
      f'avg val wells={sum(s["n_val_wells"] for s in fold_stats)/N_SPLITS:.1f} | '
      f'avg train samples={sum(s["n_train_samples"] for s in fold_stats)/N_SPLITS:,.0f}')
print('=' * 60)
print('GroupKFold configuration is valid. Safe to proceed.')
```

```python
X_all  = df[FEATURES].values
y_all  = df[TARGET].values
groups = df['Well_Name'].values

random_search = RandomizedSearchCV(
    estimator=xgb_model,
    param_distributions=xgb_param_dist,
    n_iter=N_ITER,
    scoring='r2',
    cv=GroupKFold(n_splits=N_SPLITS),
    n_jobs=1,
    random_state=RANDOM_STATE,
    verbose=2
)

print(f'Starting optimization with GroupKFold (cv={N_SPLITS})...')
start = time.time()
random_search.fit(X_all, y_all, groups=groups)
time_opt = time.time() - start

best_params   = random_search.best_params_
best_cv_score = random_search.best_score_

print(f'\nOptimization completed in {time_opt/60:.2f} min')
print(f'Best R² (GroupKFold): {best_cv_score:.4f}')
print('\nBest hyperparameters:')
for k, v in sorted(best_params.items()):
    print(f'  {k:25s}: {v}')
```

```python
# Save best parameters and cv_results
with open(MODELS_DIR / 'xgboost_best_params.json', 'w') as f:
    json.dump({'best_params': best_params, 'best_cv_score': float(best_cv_score)}, f, indent=2)

cv_results_df = pd.DataFrame(random_search.cv_results_)
cv_results_df.to_csv(MODELS_DIR / 'cv_results.csv', index=False)

print(f'Parameters saved : {MODELS_DIR}/xgboost_best_params.json')
print(f'CV results saved : {MODELS_DIR}/cv_results.csv')
print(f'  R² range       : {cv_results_df["mean_test_score"].max() - cv_results_df["mean_test_score"].min():.4f}')
```

```python
# ── Load parameters from cache (skip cell above if already exists) ───────────
with open(MODELS_DIR / 'xgboost_best_params.json', 'r') as f:
    config = json.load(f)
    best_params   = config['best_params']
    best_cv_score = config['best_cv_score']

print('best_params and best_cv_score loaded.')
for k, v in sorted(best_params.items()):
    print(f'  {k:25s}: {v}')
print(f'\nCached CV R²: {best_cv_score:.4f}')
```

## 4. LOWO Validation

```python
results_xgb = []
predictions_xgb  = []
start_lowo    = time.time()

for idx, test_well in enumerate(wells, 1):
    start_poco = time.time()

    train_df = df[df['Well_Name'] != test_well]
    test_df  = df[df['Well_Name'] == test_well]

    X_train, y_train = train_df[FEATURES].values, train_df[TARGET].values
    X_test,  y_test  = test_df[FEATURES].values,  test_df[TARGET].values

    print(f'[{idx:2d}/{len(wells)}] {test_well:25s} | Train: {len(X_train):6,} | Test: {len(X_test):5,}', end=' ')

    model = XGBRegressor(**best_params, random_state=RANDOM_STATE)
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)

    r2   = r2_score(y_test, y_pred)
    rmse = np.sqrt(mean_squared_error(y_test, y_pred))
    mae  = mean_absolute_error(y_test, y_pred)
    time_poco = time.time() - start_poco

    print(f'| R²={r2:.4f} | RMSE={rmse:.2f} | {time_poco:.1f}s')

    results_xgb.append({
        'Well_Name'       : test_well,
        'R2'              : r2,
        'RMSE'            : rmse,
        'MAE'             : mae,
        'N_samples_train' : len(X_train),
        'N_samples_test'  : len(X_test),
        'Time_seconds'    : time_poco
    })

    df_pred             = test_df[['DEPTH', 'Well_Name']].copy()
    df_pred['DT_real']  = y_test
    df_pred['DT_pred']  = y_pred
    for feat in FEATURES:
        df_pred[feat] = test_df[feat].values
    predictions_xgb.append(df_pred)

time_lowo = time.time() - start_lowo
print(f'\nLOWO completed in {time_lowo/60:.2f} min')

results_df_xgb   = pd.DataFrame(results_xgb)
df_predictions_xgb = pd.concat(predictions_xgb, ignore_index=True)

# Save
results_df_xgb.to_csv(METRICS_DIR / 'lowo_xgboost_results.csv', index=False)
df_predictions_xgb.to_csv(PREDICTIONS_DIR / 'lowo_xgboost_predictions.csv', index=False)

print(f'Metrics saved   : {METRICS_DIR}/lowo_xgboost_results.csv')
print(f'Predictions saved: {PREDICTIONS_DIR}/lowo_xgboost_predictions.csv')
```

```python
# ── Load results from cache (skip cell above if already exists) ──────────────
results_df_xgb   = pd.read_csv(METRICS_DIR / 'lowo_xgboost_results.csv')
df_predictions_xgb = pd.read_csv(PREDICTIONS_DIR / 'lowo_xgboost_predictions.csv')

print(f'Results loaded.')
print(f'  Metrics    : {results_df_xgb.shape}')
print(f'  Predictions: {df_predictions_xgb.shape}')
```

# 7. SENSITIVITY ANALYSIS — GroupKFold: n_splits=10 vs n_splits=27

```python
# ════════════════════════════════════════════════════════════════════════════════
# SENSITIVITY ANALYSIS — GroupKFold: n_splits=10 vs n_splits=27
# ════════════════════════════════════════════════════════════════════════════════
# Objective: verify whether the granularity of the internal CV (n_splits=10 vs
# full LOWO-CV with n_splits=27) significantly alters the selected hyperparameters
# and the final LOWO evaluation performance.
#
# Motivation: a reviewer may question whether n_splits=10 is sufficiently
# rigorous. This experiment provides empirical evidence that the choice of
# n_splits does not substantially affect the final result.
# ════════════════════════════════════════════════════════════════════════════════

N_SPLITS_SENSITIVITY = df['Well_Name'].nunique()  # 27 — full LOWO-CV
print(f'N_SPLITS for sensitivity analysis: {N_SPLITS_SENSITIVITY}')
```

```python
# ── GroupKFold Valditation (n_splits=27) s─────────
print('GroupKFold CONFIGURATION VALIDATION — n_splits=27')
print('=' * 60)

n_wells   = df['Well_Name'].nunique()
n_samples = len(df)

print(f'Dataset     : {n_samples:,} samples | {n_wells} wells')
print(f'N_SPLITS    : {N_SPLITS_SENSITIVITY}')
print(f'Min required: N_SPLITS <= n_wells ({N_SPLITS_SENSITIVITY} <= {n_wells})')

if N_SPLITS_SENSITIVITY > n_wells:
    raise ValueError(f'N_SPLITS={N_SPLITS_SENSITIVITY} > n_wells={n_wells}.')
else:
    print(f'[OK]  N_SPLITS={N_SPLITS_SENSITIVITY} <= n_wells={n_wells}')

X_val  = df[FEATURES].values
y_val  = df[TARGET].values
grps   = df['Well_Name'].values
gkf_27 = GroupKFold(n_splits=N_SPLITS_SENSITIVITY)

print(f'\n{"Fold":>5s}  {"Train wells":>12s}  {"Val wells":>10s}  '
      f'{"Train samples":>14s}  {"Val samples":>12s}  Val well names')
print('-' * 110)

fold_stats_27 = []
for fold, (train_idx, val_idx) in enumerate(
        gkf_27.split(X_val, y_val, groups=grps), 1):
    train_wells = set(grps[train_idx])
    val_wells   = set(grps[val_idx])
    overlap     = train_wells & val_wells
    fold_stats_27.append({
        'fold'           : fold,
        'n_train_wells'  : len(train_wells),
        'n_val_wells'    : len(val_wells),
        'n_train_samples': len(train_idx),
        'n_val_samples'  : len(val_idx),
        'overlap'        : len(overlap),
    })
    val_names = ', '.join(sorted(val_wells))
    print(f'{fold:>5d}  {len(train_wells):>12d}  {len(val_wells):>10d}  '
          f'{len(train_idx):>14,}  {len(val_idx):>12,}  {val_names}')

total_overlaps = sum(s['overlap'] for s in fold_stats_27)
print()
if total_overlaps > 0:
    print(f'[FAIL] {total_overlaps} well(s) in both train and val.')
else:
    print('[OK]  No well appears in both train and validation in any fold.')

min_val = min(s['n_val_wells'] for s in fold_stats_27)
print(f'[OK]  All folds have at least {min_val} validation well(s).')
print()
print(f'Summary: {N_SPLITS_SENSITIVITY} folds | '
      f'avg train wells={sum(s["n_train_wells"] for s in fold_stats_27)/N_SPLITS_SENSITIVITY:.1f} | '
      f'avg val wells={sum(s["n_val_wells"] for s in fold_stats_27)/N_SPLITS_SENSITIVITY:.1f} | '
      f'avg train samples={sum(s["n_train_samples"] for s in fold_stats_27)/N_SPLITS_SENSITIVITY:,.0f}')
print('=' * 60)
print('GroupKFold(n_splits=27) configuration is valid. Safe to proceed.')
```

```python
# ── Hyperparameter optimization with GroupKFold(n_splits=27) ─────────────────
X_all = df[FEATURES].values
y_all = df[TARGET].values
grps  = df['Well_Name'].values

xgb_model_27 = xgb_model
#     objective='reg:squarederror',
#     tree_method='hist',
#     random_state=42,
#     n_jobs=-1,
#     device='cuda'
# )

t0_27 = time.time()

random_search_27 = RandomizedSearchCV(
    estimator          = xgb_model,
    param_distributions= xgb_param_dist,   # same search space as n_splits=10
    n_iter             = 50,
    scoring            = 'r2',
    cv                 = GroupKFold(n_splits=N_SPLITS_SENSITIVITY),
    n_jobs             = 1,
    random_state       = RANDOM_STATE,
    verbose            = 2,
)
random_search_27.fit(X_all, y_all, groups=grps)

t_opt_27 = time.time() - t0_27

best_params_27   = random_search_27.best_params_
best_cv_score_27 = random_search_27.best_score_

print(f'CV R² (GroupKFold=27) : {best_cv_score_27:.4f}')
print(f'CV R² (GroupKFold=10) : {best_cv_score:.4f}   ← reference')
print(f'Optimization time     : {t_opt_27/60:.1f} min')
print()
print('Optimal hyperparameters — n_splits=27:')
for k, v in sorted(best_params_27.items()):
    v_ref = best_params.get(k, '—')
    match = '✅' if str(v) == str(v_ref) else '⚠️ '
    print(f'  {match} {k:25s}: {str(v):>12s}   (n_splits=10: {v_ref})')
```

```python
# ── LOWO with hyperparameters optimized via GroupKFold(n_splits=27) ──────────
import numpy as np
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error

print('LOWO — XGBoost with hyperparameters from GroupKFold(n_splits=27)')
print('=' * 70)

resultados_27 = []
t0_lowo_27    = time.time()

for idx, test_well in enumerate(sorted(df['Well_Name'].unique()), 1):
    train_df = df[df['Well_Name'] != test_well]
    test_df  = df[df['Well_Name'] == test_well]

    X_train, y_train = train_df[FEATURES].values, train_df[TARGET].values
    X_test,  y_test  = test_df[FEATURES].values,  test_df[TARGET].values

    model_27 = XGBRegressor(
        **best_params_27,
        objective   ='reg:squarederror',
        tree_method ='hist',
        random_state=RANDOM_STATE,
        device      ='cuda',
    )
    model_27.fit(X_train, y_train)
    y_pred = model_27.predict(X_test)

    r2   = r2_score(y_test, y_pred)
    rmse = np.sqrt(mean_squared_error(y_test, y_pred))
    mae  = mean_absolute_error(y_test, y_pred)

    print(f'[{idx:2d}/27] {test_well:30s} R²={r2:.4f}  RMSE={rmse:.2f}  MAE={mae:.2f}')

    resultados_27.append({
        'Well_Name': test_well,
        'R2'       : r2,
        'RMSE'     : rmse,
        'MAE'      : mae,
    })

t_lowo_27 = time.time() - t0_lowo_27
df_res_27 = pd.DataFrame(resultados_27)

print()
print(f'LOWO time: {t_lowo_27/60:.1f} min')
print(f'Total time (opt + LOWO): {(t_opt_27 + t_lowo_27)/60:.1f} min')
```
